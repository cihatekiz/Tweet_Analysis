```python
import re
import string
```

```python

#removing punctuations
def remove_punctuations(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    return text
```

```python
#removing digits
def remove_digits(text):
    for digits in string.digits:
        text=text.replace(digits,'')
    return text    
```

```python
df["cleaned_text"] = df['cleared_text'].apply(remove_punctuations)
#removing numbers
df['new_text'] = df['cleaned_text'].apply(remove_digits)
```

```python
df["new_text"]=df["new_text"].astype(str)
```

```python
def data_preprocessing(source):
    source = source.replace(' da ',' ')
    source = source.replace(' da,',', ')
    source = source.replace(" de " , " ")
    source = source.replace(" vs " , " ")
    source = source.replace(" ve " , " ")
    return source
```

```python
df["new_text"] = df['new_text'].apply(data_preprocessing)
```

```python
def y_label(source):
    empty=[]
    for i in range(len(source)):
        empty.append(source[i][1])
    return empty  
```

```python
tweet_counts=y_label(sorted_h_tweet)
```

```python
def x_label(source):
    empty=[]
    for i in range(len(source)):
        empty.append(source[i][0])
    return empty 
```

```python
from statistics import mean
import numpy as np

xs = np.array(x_label(weekday_hour_tweet), dtype=np.float64)
ys = np.array(y_label(weekday_hour_tweet), dtype=np.float64)

def best_fit_slope_and_intercept(xs,ys):
    m = (((mean(xs)*mean(ys)) - mean(xs*ys)) /
         ((mean(xs)*mean(xs)) - mean(xs*xs)))
    
    b = mean(ys) - m*mean(xs)
    
    return m, b

m, b = best_fit_slope_and_intercept(xs,ys)

print(m,b)
```

```python
regression_line = [(m*x)+b for x in xs]
```

```python
regression_line = []
for x in xs:
    regression_line.append((m*x)+b)
```
